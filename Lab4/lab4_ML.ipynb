```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Walmart.csv") 
```

```python
print("==== Task 1: Data Quality Check ====")
print("\nInfo:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nStatistics:")
print(df.describe())

print("\nDuplicates:", df.duplicated().sum())
```

```python
print("\n==== Task 2: Handle Missing Values ====")
numerical_cols = ['Weekly_Sales', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']

for col in numerical_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
print("==== Task 3: Detect & Handle Outliers ====")
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

```python
print("==== Task 4: Normalize Numerical Features ====")


minmax_scaler = MinMaxScaler()
df_minmax = df.copy()
df_minmax[numerical_cols] = minmax_scaler.fit_transform(df[numerical_cols])


zscore_scaler = StandardScaler()
df_zscore = df.copy()
df_zscore[numerical_cols] = zscore_scaler.fit_transform(df[numerical_cols])

print("Min-Max scaling applied:\n", df_minmax[numerical_cols].head())
print("\nZ-score scaling applied:\n", df_zscore[numerical_cols].head())
```

```python
print("==== Task 5: Apply PCA ====")
correlation = df[numerical_cols].corr()
print("\nCorrelation Matrix:\n", correlation)


if (correlation.abs() > 0.5).sum().sum() > len(numerical_cols):
    print("\nStrong correlation found → Applying PCA...")
    pca = PCA(n_components=2)
    pca_result = pca.fit_transform(df[numerical_cols])
    df_pca = pd.DataFrame(pca_result, columns=['PC1', 'PC2'])
    print("\nPCA Result (first 5 rows):\n", df_pca.head())

   
    plt.figure(figsize=(8,6))
    sns.scatterplot(x='PC1', y='PC2', data=df_pca)
    plt.title('PCA Scatter Plot')
    plt.xlabel('Principal Component 1 (PC1)')
    plt.ylabel('Principal Component 2 (PC2)')
    plt.grid(True)
    plt.show()
else:
    print("\nNo strong correlation → PCA not applied")
```
